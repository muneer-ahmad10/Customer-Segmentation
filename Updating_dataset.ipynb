# **Load Data**

```python
customer_data=pd.read_excel('/content/drive/MyDrive/Projects/Customer Segmentation/Data/customer_data.xlsx')
sales_data=pd.read_excel('/content/drive/MyDrive/Projects/Customer Segmentation/Data/sales_data.xlsx')
shoping_data=pd.read_excel('/content/drive/MyDrive/Projects/Customer Segmentation/Data/shopping_mall_data.xlsx')
```

## **Merging three tables**

```python
customer_data.columns
```

```python
sales_data.columns
```

```python
shoping_data.columns
```

```python
df = pd.merge(customer_data, sales_data, on='customer_id', how='inner')
merged_df=pd.merge(df, shoping_data, on='shopping_mall', how='left')
```

```python
merged_df.columns
```

```python
merged_df.sample(10)
```

## **Converting to customer level data**

```python
final_df = merged_df.groupby('customer_id').agg({
    'age': 'first',
    'gender': 'first',
    'payment_method': 'first',
    'quantity': 'sum',
    'price': 'sum',
    'invoice_no': 'count',   # number of purchases
    'store_count': 'mean'
}).reset_index()
```

```python
final_df
```

## **Renaming some Columns**

```python
final_df.rename(columns={
    'invoice_no': 'purchase_count',
    'price': 'total_spent'
}, inplace=True)
```
